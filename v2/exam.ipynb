```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

model_name = "cenkersisman/gpt2-turkish-128-token"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
device = "cuda"
```

```python
with open("text.txt", "r") as f:
    text = f.read()

text
```

```python
token_ids = tokenizer.encode(text)
token_ids
```

```python
token_ids
```

```python
import torch
a = torch.tensor(token_ids)
```

```python
context_length = 12
stride = 2
```

```python
from koRKut_dataloader import TextDataset
korkut_dataloader = TextDataset(token_ids=a, context_length=context_length, stride=stride)
korkut_dataloader.inputs[0], korkut_dataloader.targets[0]
```

```python
tokenizer.vocab_size
```

```python
from koRKut_model import koRKutModel

korkut_model = koRKutModel(vocab_size=tokenizer.vocab_size, embedding_dim=16, num_heads=4, context_length=12, num_layers=2, device=device)
```

```python
import torch
optimizer = torch.optim.Adam(korkut_model.parameters(), lr=1e-3)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
from koRKut_train import koRKutTrain

trainer = koRKutTrain(korkut_model, korkut_dataloader, optimizer, loss_fn, device=device)
trainer.train(epochs=100)
```

```python
# the capital of the united states is not london.
sentence = "Derin öğrenme"
sentence_token = tokenizer.encode(sentence)
sentence_token
```
